# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [0,2,4,6,8]
y = [0,4,16,36,64]
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, marker='o', label="Data Points", color='red')
ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()
```

Combining NumPy, Pandas and Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
arr = np.array([10,70,30,40,80,100,70,20,10,50])
df = pd.DataFrame({"Numbers": arr})
```

```python
plt.figure(figsize=(4,3))
plt.bar(df.index, df["Numbers"], color = 'purple')
plt.show()
```

Task 1 - Create a Dictionary, convert it into dataframe and then make bar chart for it

```python
dictionary = {'A': 30, 'B': 20, 'C': 50, 'D': 20, 'E': 40, 'F': 60, 'G': 100, 'H': 70, 'I': 30, 'J': 80}
df_dict = pd.DataFrame(list(dictionary.items()), columns=['Category', 'Value'])
```

```python
plt.figure(figsize=(4,3))
plt.bar(df_dict['Category'], df_dict['Value'], color='orange')
plt.title("Bar Chart from Dictionary")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```
